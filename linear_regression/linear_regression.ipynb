### Generate dataset
1. Сгенирировать выборку. Количество экземлпяров = порядковому номеру * 10, количество признаков = порядковому номеру, из них информативных  - 20% (округлять в большую сторону, минимум 1), шум - случайное число в диапозоне 1-20.

```python
from sklearn import datasets

var_num = 32
noise = random.randint(1,20)

data, target, coef = datasets.make_regression(n_features = var_num, n_informative = int(ceil(var_num / 5)), n_targets = 1, 
                                              noise = noise, coef = True, random_state = 2, n_samples = var_num * 10)
```

### Plot data
2. Вывести графики зависимости target от всех признаков.

```python
pyplot.figure(figsize(25, 35))

for plot_num in range(var_num):
    pyplot.subplot(7, 5, plot_num + 1)
    pyplot.scatter(list(map(lambda x : x[plot_num], data)), target, color = 'r')
    pyplot.xlabel('feature'+ str(plot_num))
    pyplot.ylabel('target')
```

### Linear regression model learning
3. Обучить модель линейной регрессии на всей сгенерированной выборке. 

```python
from sklearn import linear_model

regression_model = linear_model.LinearRegression()
regression_model.fit(data, target)
predictions = regression_model.predict(data)
```

### Print linear regression coefficients
4. Вывести обученную модель в виде уравнения.

```python
equation = 'y = '
equation += str(round(regression_model.intercept_, 2))

for idx, coef in enumerate(regression_model.coef_):
    equation += ' {:+.3f}*x{}'.format(coef, str(idx))

print(equation)
```

### Plot predictions

```python
pyplot.figure(figsize(10, 7))

pyplot.scatter(predictions, target, color = 'r')
pyplot.xlabel('predictions')
pyplot.ylabel('target')
```

### Calculate RMSE
5. Посчитать RMSE на всей выборке.

```python
from sklearn import metrics

mse = metrics.mean_squared_error(target, predictions)
rmse = math.sqrt(mse)
print ('RMSE =', rmse)
```
